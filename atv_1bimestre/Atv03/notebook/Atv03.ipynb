Usando a classe JOptionPane para entrada de dados, receba a nota de duas provas e de um trabalho. 

```java
import javax.swing.JOptionPane;

String aux;
float prova1, prova2, notatrabalho, mediatotal;

aux = JOptionPane.showInputDialog("Digite a nota da primeira prova");
prova1 = Float.parseFloat(aux); 

aux = JOptionPane.showInputDialog("Digite a nota da segunda prova");
prova2 = Float.parseFloat(aux);

aux = JOptionPane.showInputDialog("Digite a nota do trabalho");
notatrabalho = Float.parseFloat(aux);
```

Calcule a média 

```java
mediatotal= (prova1 + prova2 + notatrabalho) / 3; //fazendo o cálculo para calcular a media
```

Mostre a média usando o showMessageDialog da classe JOptionPane

```java
JOptionPane.showMessageDialog(null, "Á media do aluno foi " + mediatotal); 
```
